```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculateVIF(x_train_in):
    vif=pd.DataFrame()
    vif['feature']=x_train_in.columns
    vif['VIF']=[variance_inflation_factor(x_train_in.values,i) for i in range(len(x_train_in.columns))]
    vif['VIF']=round(vif['VIF'],2)
    return vif
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
def build_lm(col_to_drop):
    x_train_in=x_train.drop(col_to_drop,axis=1)
    x_test_in=x_test.drop(col_to_drop,axis=1)
    print("Model Build using ",x_train_in.columns,'\n')
    vif=calculateVIF(x_train_in)
    x_train_sm=sm.add_constant(x_train_in)
    x_test_sm=sm.add_constant(x_test_in)
    lm=sm.GLM(list(y_train_win),x_train_sm).fit()
    p_values=lm.pvalues.to_frame().rename({0:'p_values'},axis=1)
    coefficients=lm.params.to_frame().rename({0:'coefficient'},axis=1)
    merged_df=p_values.merge(vif,how='left',left_index=True,right_on='feature').merge(coefficients,how='left',left_on='feature',right_index=True)
#    merged_df=merged_df[(merged_df.VIF>5) |(merged_df.p_values>0.05)|(merged_df.VIF<-0.05)]
    merged_df=merged_df[['feature','p_values','VIF','coefficient']]
    merged_df.p_values=round(merged_df.p_values,3)
    merged_df.VIF=round(merged_df.VIF,3)
    merged_df=merged_df.sort_values(by=['p_values','VIF'],ascending=False)

    lm=LogisticRegression(class_weight='balanced')
    lm.fit(x_train_in,y_train_win)
    y_train_win_predict=lm.predict(x_train_in)

    print("accuracy_score train: %.2f" % accuracy_score(y_train_win, y_train_win_predict))
    print(merged_df)
```
